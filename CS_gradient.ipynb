```python
import CS_functions as cs
import numpy as np
from tqdm import tqdm
from matplotlib import pyplot as plt, colors as mplcolors
from scipy import fft as spfft
plt.rcParams.update({'font.size':16})
#np.set_printoptions(threshold=sys.maxsize)
```

```python
file_name = "1dmockanderrors15"
file_type = ".csv"
reduced_points = 4
regularization_coeffient = 1e-4

optlocs_file = "data\\" + file_name +"_optlocs.csv"
target, uncertainties = cs.open_dataset(file_name, file_type)
total_points = len(target)
```

```python
target_grad = np.gradient(target)
target_gradgrad = np.gradient(target_grad)
```

```python
fig, axs = plt.subplots(1, 1, figsize= (16, 8))

axs.plot(target +1)
axs.plot(target_grad)
axs.plot(target_gradgrad -1)

fig.show()
```

```python
fig, axs = plt.subplots(1, 1, figsize= (16, 8))

axs.plot(np.arctan(target) +1)
axs.plot(np.arctan(target_grad))
axs.plot(np.arctan(target_gradgrad) -1)

fig.show()
```

```python
target_dct = spfft.dct(target)
target_grad_dct = spfft.dct(target_grad)
target_gradgrad_dct = spfft.dct(target_gradgrad)


fig, axs = plt.subplots(1, 1, figsize= (16, 8))

axs.plot(np.abs(target_dct) +1)
axs.plot(np.abs(target_grad_dct))
axs.plot(np.abs(target_gradgrad_dct) -1)

fig.show()
```
